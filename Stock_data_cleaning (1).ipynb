```python
# Import libraries for pulling and estimating stock returns
import os

folder = 'Stockdata_v2'
files = os.listdir(folder)
```

```python
# Import libraries for pulling and estimating stock returns
import pandas as pd

dataframes = {}

for file in files:
    dataframes[file] = pd.read_csv(os.path.join(folder, file))
```

```python
# Estimate weekly returns of each stock based on data
for key in dataframes:
    dataframes[key]['weekly_returns_'+key[:-4]] = dataframes[key]['Close'].pct_change()
    print(dataframes[key].head())
```

```python
last_columns = {}

for key in dataframes:
    last_columns[key] = dataframes[key].iloc[:, -1]
```

```python
# Pull all stock returns together
df = pd.concat(list(last_columns.values()), axis=1)
df = df.iloc[1: , :]
print(df.head())

```

```python
# Save the stock returns in a csv file
df.to_csv('gdrive/My Drive/stock_returns_df.csv')
```
