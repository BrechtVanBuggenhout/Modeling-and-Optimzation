```python
# Minimize 
5*x1 + 2*x2 + 8*x3 + 7x4 

# constraints
 x2 + 2*x3 + 3*x4  >= 4
 4*x1 + 2*x2 - 3*x3 + 9*x4  >= 7
 2*x1 - 1*x2 + 2*x3 + 1*x4  >= 4
 x1, x2, x3, x4 >= 0
 x1, x2, x3, x4 <= 20

```

```python
n = 4
obj_coef = [5,2,8,7]
c1_coef = [0,1,2,3]
c2_coef = [4,2,-3,9]
c3_coef = [2,-1,2,1]
c1_rhs = 4
c2_rhs = 7
c3_rhs = 4
```

```python
# solve with Pyomo
# initialize concrete model
model = ConcreteModel()

# initialize dvs
model.x = Var(range(n), domain = NonNegativeReals)

# define the objective
model.Objective = Objective(expr = sum(obj_coef[i]*model.x[i] for i in range(n)), sense = minimize)

# specify constraints
model.Constraint1 = Constraint(expr = sum(c1_coef[i]*model.x[i] for i in range(n)) >= c1_rhs)
model.Constraint2 = Constraint(expr = sum(c2_coef[i]*model.x[i] for i in range(n)) >= c2_rhs)
model.Constraint3 = Constraint(expr = sum(c3_coef[i]*model.x[i] for i in range(n)) >= c3_rhs)
```

```python
# solve model
opt = SolverFactory('glpk', executable='/usr/bin/glpsol')

results = opt.solve(model, tee = True)
```

```python
# print relevant values
for i in range(n):
  print(f"x{i} = {model.x[i]()}")
print("Obj* = ", model.Objective())
```
