```python
# input parameters
unit_price = [125,345,99,380,235]
volume = [8400,4200,10950,4000,3850]
unit_cost = [62,188,52,225,109]
sales_com_factor = 0.03
num_products = len(unit_price)
```

```python
# GP calculations method 1 = for loop
gross_sales = []
cogs = []
gross_profit = []
for i in range(num_products):
  gross_sales.append(unit_price[i]*volume[i])
  cogs.append(unit_cost[i]*volume[i])
  gross_profit.append(gross_sales[i]-cogs[i])
print("gross profit = ",gross_profit)
total_GP = sum(gross_profit)
print("total gross profit =",total_GP)
```

```python
# expense calculations
sales_com = [sales_com_factor*gross_sales[i] for i in range(num_products)]
print("sales comission =", sales_com)
total_sales_com = sum(sales_com)
print("total sales com =", total_sales_com)
```

```python
# net earnings output
net_earnings = total_GP - total_sales_com
print("net earnings=", net_earnings)
```
