```python
# inputs
num_years = 5
Sales_Price = 80
Sales_volume = 120000
Var_Prod_cost = 50
fixed_prod_cost = 32000

sales_pr_growth = 0.05
sales_vol_growth = 0.075
var_prod_growth = 0.02
sga_rate = 0.08
tax_rate = 0.34
disc_factor = 0.1
```

```python
# Creating the model
# revenue calculations
base_sales_units = [(1+sales_vol_growth)** i for i in range(num_years)] # calculating % growth rate for each year
base_sales_price = [(1+sales_pr_growth)** i for i in range(num_years)] # calculating % growth rate for each year
sales_dollars = [Sales_volume*base_sales_units[i]*(Sales_Price*base_sales_price[i]) for i in range(num_years)]
print("sales revenue", sales_dollars)
# Product cost calculation
base_var_cost = [(1+var_prod_growth)**i for i in range(num_years)]
var_cost = [(Var_Prod_cost*base_var_cost[i]) for i in range(num_years)]
var_cost_total = [var_cost[i]*base_sales_units[i]*Sales_volume for i in range(num_years)]
prod_cost = [var_cost_total[i] + fixed_prod_cost for i in range(num_years)]
print("Product costs", prod_cost)
#SGA expense
sga_exp = [prod_cost[i]*sga_rate for i in range(num_years)]
print('SG&A Expense', sga_exp)
# income calculation
income = [sales_dollars[i] - prod_cost[i]-sga_exp[i] for i in range(num_years)]
print('income before tax', income)
# income after tax
inc_aft_tax = [income[i]*(1-0.34) for i in range(num_years)]
print("earnings after tax: ", inc_aft_tax)
```

```python
# caclulating npv factors
npv_factor = [1/(1+disc_factor)** (i+1) for i in range(num_years)]
npv_factor
```

```python
# Calculating NPV of earnings after tax
npv_business = [inc_aft_tax[i]*npv_factor[i] for i in range(num_years)]
npv_business
sum_npv_business = sum(npv_business)
sum_npv_business
print("NPV of earnings after tax =", sum_npv_business)
```
