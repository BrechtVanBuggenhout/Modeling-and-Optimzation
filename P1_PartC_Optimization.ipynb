```python
# Minimize 
ùë•^2 ‚àí ùë•ùë¶ + ùë¶^2 ‚àí 3ùë• ‚àí 1.5ùë¶

# constraints
‚àí1 ‚â§ ùë• ‚â§ 2
0 ‚â§ ùë¶ ‚â§ 3

```

```python
# solve with Pyomo
# initialize concrete model
model = ConcreteModel()

# initialize dvs
model.x = Var(domain = Reals, bounds = (-1,2))
model.y = Var(domain = Reals, bounds = (0,3))

# define the objective
model.Objective = Objective(expr = model.x**2 - model.x*model.y + model.y**2 - 3*model.x - 1.5*model.y, sense = minimize)

```

```python
# solve model
opt = SolverFactory('ipopt', executable='/content/ipopt')

results = opt.solve(model, tee = True)
```

```python
# print relevant values
print("x = ", model.x())
print("y = ", model.y())
print("objective = ", model.Objective())
```
