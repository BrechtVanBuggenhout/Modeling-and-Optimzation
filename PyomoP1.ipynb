minimize f(x) = 4x^4 - 3x^3 - 9x + 20


```python
def my_function(x):
  y = 4*x**4 - 3*x**3 - 9*x + 20
  return y
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,1000)

y = [my_function(x[i]) for i in range(1000)]
```

```python
# solve with Pyomo
# initialize concrete model
model = ConcreteModel()
# initialize dvs
model.x = Var(domain = Reals)
# define the objective
model.Objective = Objective(expr = 4*model.x**4 - 3*model.x**3 - 9*model.x + 20, sense = minimize)
```

```python
# solve model
opt = SolverFactory('ipopt', executable = '/content/ipopt')
results = opt.solve(model, tee = True)
```

```python
# print relevant values
print("x* = ",model.x())
print("obj* = ", model.Objective())
```
