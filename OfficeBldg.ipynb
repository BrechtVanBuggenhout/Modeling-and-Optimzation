```python
# Inputs
area_bldg = 180000 # sq ft
rent_rate = 15
m = -0.05
b = 1.5
mortgage_pmt = 1500000
op_cost_growth = 0.12
op_cost = 1.2
tax_rate = 0.34
heating_rate = 0.2

year = [0,1,2,3,4]
num_years = len(year)
```

```python
# % occupied, avg operating costs, and heating
pct_occupied = rent_rate*m+b
pct_occupied
avg_operating_costs = area_bldg*op_cost
avg_operating_costs
heating = avg_operating_costs*heating_rate*pct_occupied
heating
```

```python
# calculation of revenue, operating costs, and mortgage
revenue = []
operating_costs = []
mortgage_payment = []

for i in range(num_years):
  revenue.append(area_bldg*rent_rate*pct_occupied)
  operating_costs.append((avg_operating_costs+heating)*(1+op_cost_growth)**year[i])
  mortgage_payment.append(mortgage_pmt)

print("revenue = ", revenue)
print("operating costs = " ,operating_costs)
print("mortgage payment = ", mortgage_payment)
```

```python
# calculation of income before income taxes
inc_before_tax = []

for i in range(num_years):
  inc_before_tax.append(revenue[i] - operating_costs[i] - mortgage_payment[i])

print(" income before taxes = ", inc_before_tax)
```

```python
# calculation of taxes
taxes = []

for i in range(num_years):
  taxes.append(inc_before_tax[i]*tax_rate)

print("taxes =", taxes)
```

```python
# Income after tax
income_after_tax = []

for i in range(num_years):
  income_after_tax.append(inc_before_tax[i]-taxes[i])

print("income after tax = ", income_after_tax)

total_net_income = sum(income_after_tax)
print("Total Net Income = ", total_net_income)
```
