# AdBudget Model 2:30pm class


```python
# Inputs
unit_cost = 25 #dollars
sales_price = 40 #dollars
seasonality_factor = [0.9,1.1,0.8,1.2] #for q1, q2, q3, q4
salesforce_cost = [8000,8000,9000,9000]
overhead_rate = 0.15 
k = 35
b = 3000
ad_spend = [10000,10000,10000,10000]

num_quarters = len(ad_spend)
```

```python
# GP Calculation
sales_quantity = [k*seasonality_factor[i]*((b+ad_spend[i])**.5) for i in range(num_quarters)]
sales_quantity
cogs = [unit_cost*sales_quantity[i] for i in range(num_quarters)]
cogs
sales_rev = [sales_price*sales_quantity[i] for i in range(num_quarters)]
sales_rev
gross_profit = [sales_rev[i]-cogs[i] for i in range(num_quarters)]
print("gross profit =", gross_profit)
total_gross_profit = sum(gross_profit)
print("total gross profit = ",total_gross_profit)
```

```python
# Operating Expenses calcualtion
overhead = []
operating_expenses = []

for i in range(num_quarters):
  overhead.append(overhead_rate*sales_rev[i])
  operating_expenses.append(overhead[i]+salesforce_cost[i]+ad_spend[i])

print("overhead =",overhead)
print("operating expenses = ", operating_expenses)

total_operating_expenses = sum(operating_expenses)
print("total operating expenses = ", total_operating_expenses)
```

```python
# Annual operating profit
total_op_profit = total_gross_profit-total_operating_expenses
print("total operating profit =" ,total_op_profit)
```
